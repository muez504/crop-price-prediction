# let's remove the outlayers for each price 25%<values<75% out of this range

```python
#IQR
Q1=np.percentile(data2['Producer Price'], 25,interpolation="midpoint")
Q3=np.percentile(data2['Producer Price'], 75,interpolation="midpoint")
IQR=Q3-Q1
```

```python
#Upper bound
upper=np.where(data2['Producer Price']>=(Q3+1.5*IQR))
#Lower bound
lower=np.where(data2['Producer Price']<=(Q1-1.5*IQR))
```

```python
print(upper[0],lower[0])
```

```python
#...Removing the outliers...
data2.drop(upper[0], inplace=True)
data2.drop(lower[0], inplace=True)

print("New Shape: ", data2.shape)
```
